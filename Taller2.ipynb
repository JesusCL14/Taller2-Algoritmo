## Filtros y análisis de datos
Aquí ocupamos la librería Pandas para leer el archivo CSV y hacer consultas. Armamos funciones para filtrar la información, por ejemplo, para buscar estudiantes por género o ver cómo influye la educación de los padres en las notas. Tratamos de hacer el código reutilizable y tuvimos en cuenta los datos nulos (NaN) para que el programa no tire error si llega a faltar información en alguna fila.

```python
# Creamos la función reutilizable
def obtener_top_mujeres(dataframe, cantidad):
    # 1. Filtramos para quedarnos solo con las mujeres
    df_mujeres = dataframe[dataframe['gender'] == 'Female']

    # 2. Ordenamos por nota final ('final_score') de mayor a menor (ascending=False)
    df_ordenado = df_mujeres.sort_values(by='final_score', ascending=False)

    # 3. Cortamos la lista para mostrar solo la cantidad que pediste
    return df_ordenado.head(cantidad)

# ==========================================
# ⚙️ PANEL DE CONTROL (Modifica esto)
# ==========================================

CANTIDAD_A_MOSTRAR = 5  # <--- ¡Cambia este número por 10, 50 o el que quieras!

# ==========================================

# Ejecutamos la función con nuestra variable 'df'
resultado = obtener_top_mujeres(df, CANTIDAD_A_MOSTRAR)

# Mostramos la tabla final
resultado
```

```python
# Creamos una nueva función reutilizable
def obtener_peores_notas(dataframe, cantidad):
    # 1. Ordenamos por nota final ('final_score') de menor a mayor (ascending=True)
    # Al ser True, los números más bajos (peores notas) quedarán arriba
    df_ordenado = dataframe.sort_values(by='final_score', ascending=True)

    # 2. Cortamos la lista para mostrar solo la cantidad que pediste
    return df_ordenado.head(cantidad)

# ==========================================
# ⚙️ PANEL DE CONTROL (Modifica esto)
# ==========================================

CANTIDAD_A_MOSTRAR_PEORES = 5  # <--- ¡Cambia este número por el que quieras!

# ==========================================

# Ejecutamos la función con nuestra variable 'df'
resultado_peores = obtener_peores_notas(df, CANTIDAD_A_MOSTRAR_PEORES)

# Mostramos la tabla final
resultado_peores
```

```python
# Creamos la función reutilizable para el criterio del Mérito (o cualquier otro nivel)
def filtrar_por_educacion_y_mejores_notas(dataframe, nivel_educacion, cantidad):

    # 1. Filtramos por el nivel educativo
    # Si escribes "NaN", usamos una herramienta especial de Pandas para buscar datos vacíos (.isna)
    if nivel_educacion == "NaN":
        df_filtrado = dataframe[dataframe['parent_education'].isna()]
    else:
        # Si es un texto normal ('High School', 'Bachelor', etc.), filtramos exacto
        df_filtrado = dataframe[dataframe['parent_education'] == nivel_educacion]

    # 2. Ordenamos a ese grupo filtrado por su nota final (de mayor a menor)
    df_ordenado = df_filtrado.sort_values(by='final_score', ascending=False)

    # 3. Retornamos solo la cantidad de estudiantes que pediste
    return df_ordenado.head(cantidad)

# ==========================================
# ⚙️ PANEL DE CONTROL (Modifica esto)
# ==========================================

# Opciones que puedes escribir aquí: 'High School', 'Bachelor', 'Master', 'PhD', 'NaN'
NIVEL_EDUCATIVO = 'High School'

# ¿Cuántos estudiantes quieres ver?
CANTIDAD_A_MOSTRAR_MERITO = 5

# ==========================================

# Ejecutamos la función
resultado_merito = filtrar_por_educacion_y_mejores_notas(df, NIVEL_EDUCATIVO, CANTIDAD_A_MOSTRAR_MERITO)

# Mostramos la tabla final
resultado_merito
```

```python
# Creamos la función reutilizable para el criterio de acceso a internet
def filtrar_por_internet_y_mejores_notas(dataframe, tiene_internet, cantidad):

    # 1. Filtramos por si tienen o no acceso a internet ('Yes' o 'No')
    df_filtrado = dataframe[dataframe['internet_access'] == tiene_internet]

    # 2. Ordenamos a ese grupo por su nota final de mayor a menor
    df_ordenado = df_filtrado.sort_values(by='final_score', ascending=False)

    # 3. Retornamos la cantidad exacta que pediste
    return df_ordenado.head(cantidad)

# ==========================================
# ⚙️ PANEL DE CONTROL (Modifica esto)
# ==========================================

# Opciones: 'No' (Sin internet) o 'Yes' (Con internet)
ACCESO_A_INTERNET = 'Yes'

# ¿Cuántos estudiantes quieres ver?
CANTIDAD_A_MOSTRAR_INTERNET = 5

# ==========================================

# Ejecutamos la función
resultado_internet = filtrar_por_internet_y_mejores_notas(df, ACCESO_A_INTERNET, CANTIDAD_A_MOSTRAR_INTERNET)

# Mostramos la tabla final
resultado_internet
```

```python
# Creamos la función reutilizable para ver a los más estudiosos
def obtener_mas_estudiosos(dataframe, cantidad):

    # Ordenamos por dos columnas a la vez:
    # 1° 'study_hours_per_week' (Horas de estudio)
    # 2° 'final_score' (Nota final, para desempatar)
    # Ambas de mayor a menor (ascending=[False, False])
    df_ordenado = dataframe.sort_values(
        by=['study_hours_per_week', 'final_score'],
        ascending=[False, False]
    )

    # Cortamos la lista para mostrar la cantidad que pediste
    return df_ordenado.head(cantidad)

# ==========================================
# ⚙️ PANEL DE CONTROL (Modifica esto)
# ==========================================

# ¿Cuántos de los estudiantes "más estudiosos" quieres ver?
CANTIDAD_A_MOSTRAR_ESTUDIOSOS = 5  # Lo puse en 10 para que veas más variedad, ¡cámbialo si quieres!

# ==========================================

# Ejecutamos la función
resultado_estudiosos = obtener_mas_estudiosos(df, CANTIDAD_A_MOSTRAR_ESTUDIOSOS)

# Para que se vea más limpio, vamos a mostrar solo las columnas que nos interesan
columnas_interes = ['student_id', 'study_hours_per_week', 'final_score', 'passed']
resultado_estudiosos[columnas_interes]
```

## Comparación de Algoritmos
Para esta parte pasamos los datos a listas normales de Python para probar nuestros propios algoritmos y comparar los tiempos. Nos basamos en los algoritmos del apunte de la clase:

- Algoritmo Burbuja: Lo implementamos para tener la base del algoritmo lento. Tiene complejidad $O(n^2)$ porque va comparando todo de a pares y se demora bastante.
- Algoritmo Quicksort: Este es nuestro algoritmo optimizado. Ocupa la lógica de dividir y vencerás con una complejidad de $O(n \log n)$, por lo que debería demorarse mucho menos.
- Función nativa de Python: Agregamos la función sorted() que ya viene lista en Python (que usa Timsort) solo para ver la diferencia de tiempo frente a los códigos que hicimos nosotros a mano.

```python
import pandas as pd
import time

# ==========================================
# 1. CARGA DE DATOS
# ==========================================
print("Cargando datos de los 500 estudiantes...")
df = pd.read_csv('student_performance.csv')
lista_estudiantes = df.to_dict('records') # Convertimos a lista para los algoritmos manuales
print("¡Datos cargados con éxito!\n")

# ==========================================
# 2. DEFINICIÓN DE LOS ALGORITMOS
# ==========================================

# --- ALGORITMO 1: BUBBLE SORT (El lento) ---
def bubble_sort(lista, crit1, crit2):
    arr = lista.copy()
    n = len(arr)
    for i in range(n):
        intercambio = False
        for j in range(0, n - i - 1):
            # Orden descendente (de mayor a menor)
            if arr[j][crit1] < arr[j+1][crit1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                intercambio = True
            elif arr[j][crit1] == arr[j+1][crit1]:
                if arr[j][crit2] < arr[j+1][crit2]:
                    arr[j], arr[j+1] = arr[j+1], arr[j]
                    intercambio = True
        if not intercambio:
            break
    return arr

# --- ALGORITMO 2: QUICKSORT (El rápido de tu PPT) ---
def quicksort(lista, crit1, crit2):
    if len(lista) <= 1:
        return lista

    # Elegimos el pivote (el elemento del medio)
    pivote = lista[len(lista) // 2]
    izquierda, medio, derecha = [], [], []

    for elemento in lista:
        # Como ordenamos de MAYOR a MENOR, los más grandes van a la izquierda
        if elemento[crit1] > pivote[crit1]:
            izquierda.append(elemento)
        elif elemento[crit1] < pivote[crit1]:
            derecha.append(elemento)
        else:
            # Desempate con el Criterio 2
            if elemento[crit2] > pivote[crit2]:
                izquierda.append(elemento)
            elif elemento[crit2] < pivote[crit2]:
                derecha.append(elemento)
            else:
                medio.append(elemento)

    return quicksort(izquierda, crit1, crit2) + medio + quicksort(derecha, crit1, crit2)

# --- ALGORITMO 3: TIMSORT (El nativo de Python/Pandas) ---
def timsort(lista, crit1, crit2):
    # La función sorted() de Python usa Timsort por debajo
    # Le pasamos una "llave" para que ordene por crit1 y luego crit2, al revés (mayor a menor)
    return sorted(lista, key=lambda x: (x[crit1], x[crit2]), reverse=True)


# ==========================================
# 3. LA GRAN CARRERA (Prueba de Eficiencia)
# ==========================================
print("🏁 ¡INICIA LA CARRERA DE ORDENAMIENTO! 🏁")
print("-" * 45)

criterio_principal = 'final_score'
criterio_secundario = 'study_hours_per_week'

# Prueba 1: Bubble Sort
inicio = time.time()
resultado_bubble = bubble_sort(lista_estudiantes, criterio_principal, criterio_secundario)
tiempo_bubble = time.time() - inicio
print(f"🐌 Bubble Sort tardó : {tiempo_bubble:.6f} segundos")

# Prueba 2: Quicksort
inicio = time.time()
resultado_quick = quicksort(lista_estudiantes, criterio_principal, criterio_secundario)
tiempo_quick = time.time() - inicio
print(f"⚡ Quicksort tardó   : {tiempo_quick:.6f} segundos")

# Prueba 3: Timsort
inicio = time.time()
resultado_tim = timsort(lista_estudiantes, criterio_principal, criterio_secundario)
tiempo_tim = time.time() - inicio
print(f"🚀 Timsort tardó     : {tiempo_tim:.6f} segundos")

print("-" * 45)
print("Verificación de que los 3 dieron el mismo mejor estudiante:")
print(f"Mejor con Bubble : ID {resultado_bubble[0]['student_id']} - Nota: {resultado_bubble[0]['final_score']}")
print(f"Mejor con Quick  : ID {resultado_quick[0]['student_id']} - Nota: {resultado_quick[0]['final_score']}")
print(f"Mejor con Timsort: ID {resultado_tim[0]['student_id']} - Nota: {resultado_tim[0]['final_score']}")
```
